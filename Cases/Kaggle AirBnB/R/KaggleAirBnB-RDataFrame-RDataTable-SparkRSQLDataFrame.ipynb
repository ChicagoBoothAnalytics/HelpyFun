# Introduction

This tutorial uses data from [__AirBnB's 2015 recruiting competition on Kaggle.com__](https://www.kaggle.com/c/airbnb-recruiting-new-user-bookings) to demo the following tools:

- __`data.frame`__ and its enhanced cousin [__`data.table`__](https://cran.r-project.org/web/packages/data.table/index.html), tools in `R` for creating and manipulating `SQL`-like tabular data;

- [__`SparkR DataFrame`__ & __`SparkR SQL`__](http://spark.apache.org/docs/latest/sql-programming-guide.html), data-frame and `SQL` features shipped as part of large-scale distributed in-memory computing solution [__`Apache Spark`__](http://spark.apache.org).

To enable faster data import and demonstrating `SQL`-related capabilities, the `.CSV` data files provided on Kaggle have been uploaded to a `PostgreSQL` database hosted on Amazon Web Services (AWS) Relational Database Service (RDS) under Chicago Booth Analytics's account.

We will also showcase visualization packages that are particularly well-integrated with data frames, especially __`GGplot2`__, although visualization is not the focus of this tutorial and will be covered more extensively in other tutorials.

Let's convert certain string columns to categorical factors, and convert `timestamp_first_active` to `POSIXlt` date-times:

```r
# let's create a custom parser of character strings to POSIXlt date-times
converter_from_str_to_POSIXlt <-
    function(from) {
        s <- regmatches(from, regexpr('[0-9]+', from, perl=TRUE))
        as.POSIXlt(
            paste(
                substr(s, 1, 4), '-', substr(s, 5, 6), '-', substr(s, 7, 8), ' ',
                substr(s, 9, 10), ':', substr(s, 11, 12), ':', substr(s, 13, 14), sep=''),
            format='%Y-%m-%d %H:%M:%S')
    }

setClass('POSIXlt_DateTime')
setAs('character', 'POSIXlt_DateTime', converter_from_str_to_POSIXlt)
```

```r
users_df$gender <- factor(users_df$gender)
users_df$signup_method <- factor(users_df$signup_method)
users_df$signup_flow <- factor(users_df$signup_flow)
users_df$language <- factor(users_df$language)
users_df$affiliate_channel <- factor(users_df$affiliate_channel)
users_df$affiliate_provider <- factor(users_df$affiliate_provider)
users_df$first_affiliate_tracked <- factor(users_df$first_affiliate_tracked)
users_df$signup_app <- factor(users_df$signup_app)
users_df$first_device_type <- factor(users_df$first_device_type)
users_df$first_browser <- factor(users_df$first_browser)
users_df$country_destination <- factor(users_df$country_destination)

users_df$timestamp_first_active <- as(users_df$timestamp_first_active, 'POSIXlt_DateTime')

cat("'users' data.frame column data types after conversion:")
sapply(users_df, class)

cat("'users' data.frame summary after conversion:")
summary(users_df)
```

```r
# let's clean up by producing a new 'users' table
# through a SQL SELECT statement
# using the various findings above
users_ddf <-
    sql(sqlContext=sqlc,
        sqlQuery=
            "SELECT
                id,
                date_account_created AS acct_crt_date,
                YEAR(date_account_created) AS acct_crt_yr,
                MONTH(date_account_created) AS acct_crt_mth,
                DAY(date_account_created) AS acct_crt_day,
                CAST(DATE_FORMAT(date_account_created, 'u') AS INT) AS acct_crt_day_of_wk_idx,
                CASE WHEN date_account_created IS NULL THEN NULL
                     ELSE CONCAT(DATE_FORMAT(date_account_created, 'u'), ': ',
                                 DATE_FORMAT(date_account_created, 'E'))
                     END
                     AS acct_crt_day_of_wk,
                CAST(
                    CONCAT(
                        SUBSTR(timestamp_first_active, 1, 4),
                        '-',
                        SUBSTR(timestamp_first_active, 5, 2),
                        '-',
                        SUBSTR(timestamp_first_active, 7, 2),
                        ' ',
                        SUBSTR(timestamp_first_active, 9, 2),
                        ':',
                        SUBSTR(timestamp_first_active, 11, 2),
                        ':',
                        SUBSTR(timestamp_first_active, 13, 2))
                    AS TIMESTAMP)
                    AS fst_actv_ts,
                date_first_booking AS fst_bkg_date,
                YEAR(date_first_booking) AS fst_bkg_yr,
                MONTH(date_first_booking) AS fst_bkg_mth,
                DAY(date_first_booking) AS fst_bkg_day,
                CAST(DATE_FORMAT(date_first_booking, 'u') AS INT) AS fst_bkg_day_of_wk_idx,
                CASE WHEN date_first_booking IS NULL THEN NULL
                     ELSE CONCAT(DATE_FORMAT(date_first_booking, 'u'), ': ',
                                 DATE_FORMAT(date_first_booking, 'E'))
                     END
                     AS fst_bkg_day_of_wk,
                CASE WHEN age > 1000 THEN 2015 - age
                     ELSE age
                     END
                     AS age,
                CASE WHEN gender = '-unknown-' THEN NULL
                     ELSE gender
                     END
                     AS gender,
                language,
                signup_method,
                signup_flow,
                signup_app,
                first_device_type,
                CASE WHEN first_browser = '-unknown-' THEN NULL
                     ELSE first_browser
                     END
                     AS first_browser,
                affiliate_channel,
                affiliate_provider,
                CASE WHEN first_affiliate_tracked = 'null' THEN NULL
                     ELSE first_affiliate_tracked
                     END
                     AS first_affiliate_tracked,
                country_destination
            FROM
                users")
cache(users_ddf)
registerTempTable(users_ddf, 'users')   # this re-registration updates the 'users' table in the SQL Context 

users_ddf <-
    sql(sqlContext=sqlc,
        sqlQuery=
            paste(  
                "SELECT",
                    paste(setdiff(columns(users_ddf), 'age'), collapse=', '), ",
                    YEAR(fst_actv_ts) AS fst_actv_yr,
                    MONTH(fst_actv_ts) AS fst_actv_mth,
                    DAY(fst_actv_ts) AS fst_actv_day,
                    CAST(FROM_UNIXTIME(CAST(fst_actv_ts AS BIGINT), 'u') AS INT) AS fst_actv_day_of_wk_idx,
                    CASE WHEN fst_actv_ts IS NULL THEN NULL
                         ELSE CONCAT(FROM_UNIXTIME(CAST(fst_actv_ts AS BIGINT), 'u'), ': ',
                                     FROM_UNIXTIME(CAST(fst_actv_ts AS BIGINT), 'E'))
                         END
                         AS fst_actv_day_of_wk,
                    HOUR(fst_actv_ts) AS fst_actv_hr,
                    CASE WHEN age BETWEEN", min_age_cutoff, "AND", max_age_cutoff, "THEN age
                         ELSE NULL
                         END
                         AS age
                FROM
                    users"))
cache(users_ddf)
registerTempTable(users_ddf, 'users')   # this re-registration updates the 'users' table in the SQL Context

cat("'users' table (SparkR SQL Distributed DataFrame) schema after cleaning:\n")
printSchema(users_ddf)
```

```r
age_gender_destination_stats_ddf = sqlc.sql(
    "SELECT \
        age_bucket, \
        CASE WHEN age_bucket = '100+' THEN 99.5 \
             ELSE CAST(REGEXP_EXTRACT(age_bucket, '([0-9]*)-([0-9]*)', 1) AS DOUBLE) - 0.5 \
             END \
             AS age_from, \
        CASE WHEN age_bucket = '100+' THEN %i + 0.5 \
             ELSE CAST(REGEXP_EXTRACT(age_bucket, '([0-9]*)-([0-9]*)', 2) AS DOUBLE) + 0.5 \
             END \
             AS age_to, \
        gender, \
        country_destination, \
        population_in_thousands \
    FROM \
        age_gender_destination_stats \
    ORDER BY \
        age_from, \
        gender, \
        country_destination"
    % max_age_cutoff)\
    .cache()
age_gender_destination_stats_ddf.registerTempTable('age_gender_destination_stats')

age_gender_destination_stats_ddf.show(NB_EXAMPLES_TO_SHOW)
```

```r
%%sql

SELECT
    sub_query_2.*,
    countries.destination_language AS dest_lang,
    countries.distance_km AS dist_km,
    countries.destination_km2 AS dest_area_km2
FROM

    (SELECT
        age_bucket,
        RANK() OVER desc_nb_1st_bkgs_by_age_bucket AS dest_rank,
        dest,
        nb_1st_bkgs_2014,
        nb_1st_bkgs_2014 / SUM(nb_1st_bkgs_2014) OVER by_age_bucket AS prop_in_bucket_nb_bkgs,
        nb_1st_bkgs_2014 / nb_all_bkgs_2015 AS prop_in_all_bkgs
    FROM

        (SELECT
            age_buckets.age_bucket AS age_bucket,
            users_clean.country_destination AS dest,
            COUNT(DISTINCT users_clean.id) AS nb_1st_bkgs_2014,
            age_dest_stats.nb_all_bkgs_2015 AS nb_all_bkgs_2015
        FROM
            users_clean
                LEFT JOIN age_buckets
                    ON users_clean.age BETWEEN age_buckets.age_from AND age_buckets.age_to
                LEFT JOIN

                    (SELECT
                        age_bucket,
                        country_destination AS dest,
                        1000 * SUM(population_in_thousands) AS nb_all_bkgs_2015
                    FROM
                        age_gender_destination_stats
                    GROUP BY
                        age_bucket,
                        country_destination) age_dest_stats

                    ON users_clean.country_destination = age_dest_stats.dest AND
                       age_buckets.age_bucket = age_dest_stats.age_bucket

        WHERE
            users_clean.country_destination NOT IN ('US', 'other', 'NDF') AND
            users_clean.fst_bkg_yr = 2014
        GROUP BY
            age_buckets.age_bucket,
            users_clean.country_destination,
            nb_all_bkgs_2015) sub_query_1

    WINDOW
        by_age_bucket AS
            (PARTITION BY age_bucket),
        desc_nb_1st_bkgs_by_age_bucket AS
            (PARTITION BY age_bucket
             ORDER BY nb_1st_bkgs_2014 DESC)) sub_query_2

    LEFT JOIN countries
          ON sub_query_2.dest = countries.country_destination
WHERE
    sub_query_2.dest_rank <= 3
ORDER BY
    age_bucket,
    dest_rank
```

```r
%%sql

-- augment the above table with zeros for Destinations with no first bookings in a Month
DROP TABLE IF EXISTS nb_1st_bkgs_incl_zeros_by_mth_by_dest;

SELECT
    all_mths_and_foreign_dests.start_of_mth AS start_of_mth,
    all_mths_and_foreign_dests.dest AS dest,
    COALESCE(nb_1st_bkgs_by_mth_by_dest.nb_1st_bkgs_in_mth, 0) AS nb_1st_bkgs_in_mth
INTO
    TEMP TABLE nb_1st_bkgs_incl_zeros_by_mth_by_dest
FROM
    nb_1st_bkgs_by_mth_by_dest

    RIGHT JOIN

    (SELECT
        all_mths.start_of_mth AS start_of_mth,
        foreign_dests.column1 AS dest
    FROM

        (SELECT
            DISTINCT start_of_mth
        FROM
            nb_1st_bkgs_by_mth_by_dest) all_mths

        CROSS JOIN

        (VALUES ('AU'), ('CA'), ('DE'), ('ES'), ('FR'),
                ('GB'), ('IT'), ('NL'), ('PT')) foreign_dests) all_mths_and_foreign_dests

    ON nb_1st_bkgs_by_mth_by_dest.start_of_mth = all_mths_and_foreign_dests.start_of_mth AND
       nb_1st_bkgs_by_mth_by_dest.dest = all_mths_and_foreign_dests.dest
ORDER BY
    start_of_mth,
    dest;

SELECT
    *
FROM
    nb_1st_bkgs_incl_zeros_by_mth_by_dest
```

```r
%%sql

-- compute rankings among Destinations by Month
SELECT
    start_of_mth,
    dest_rank_in_mth,
    LAG(dest_rank_in_mth, 1, NULL) OVER inc_mth_by_dest - dest_rank_in_mth AS dest_rank_rise_vs_last_mth,
    dest,
    nb_1st_bkgs_in_mth
FROM
    (SELECT
        start_of_mth,
        RANK() OVER desc_nb_1st_bkgs_by_mth AS dest_rank_in_mth,
        dest,
        nb_1st_bkgs_in_mth
    FROM
        nb_1st_bkgs_incl_zeros_by_mth_by_dest
    WINDOW
        desc_nb_1st_bkgs_by_mth AS
            (PARTITION BY start_of_mth
             ORDER BY nb_1st_bkgs_in_mth DESC)) sub_query
WINDOW
    inc_mth_by_dest AS
        (PARTITION BY dest
         ORDER BY start_of_mth)
ORDER BY
    start_of_mth,
    dest_rank_in_mth
```

```r
print('The above table augmented with 0s for Destinations without any first bookings for a certain Month:\n(first %i rows only)\n'
      % NB_EXAMPLES_TO_SHOW)

nb_1st_bkgs_incl_zeros_by_mth_by_dest_ddf = sqlc.sql(
    "SELECT \
        all_mths_and_foreign_dests.start_of_mth AS start_of_mth, \
        all_mths_and_foreign_dests.dest AS dest, \
        COALESCE(nb_1st_bkgs_by_mth_by_dest.nb_1st_bkgs_in_mth, 0) AS nb_1st_bkgs_in_mth \
    FROM \
        nb_1st_bkgs_by_mth_by_dest \
        \
        RIGHT JOIN \
        \
        (SELECT \
            all_mths.start_of_mth AS start_of_mth, \
            foreign_dests.dest AS dest \
        FROM \
            \
            (SELECT \
                DISTINCT start_of_mth \
            FROM \
                nb_1st_bkgs_by_mth_by_dest) all_mths \
            \
            CROSS JOIN \
            \
            (SELECT \
                DISTINCT dest \
            FROM \
                nb_1st_bkgs_by_mth_by_dest) foreign_dests) all_mths_and_foreign_dests \
            \
        ON nb_1st_bkgs_by_mth_by_dest.start_of_mth = all_mths_and_foreign_dests.start_of_mth AND \
           nb_1st_bkgs_by_mth_by_dest.dest = all_mths_and_foreign_dests.dest \
    ORDER BY \
        start_of_mth, \
        dest")\
    .cache()

nb_1st_bkgs_incl_zeros_by_mth_by_dest_ddf.registerTempTable('nb_1st_bkgs_incl_zeros_by_mth_by_dest')

nb_1st_bkgs_incl_zeros_by_mth_by_dest_ddf.show(NB_EXAMPLES_TO_SHOW)
```
