# Introduction

This tutorial uses data from [__AirBnB's 2015 recruiting competition on Kaggle.com__](https://www.kaggle.com/c/airbnb-recruiting-new-user-bookings) to demo the following tools:

- __`data.frame`__ and its enhanced cousin [__`data.table`__](https://cran.r-project.org/web/packages/data.table/index.html), tools in `R` for creating and manipulating `SQL`-like tabular data;

- [__`SparkR DataFrame`__ & __`SparkR SQL`__](http://spark.apache.org/docs/latest/sql-programming-guide.html), data-frame and `SQL` features shipped as part of large-scale distributed in-memory computing solution [__`Apache Spark`__](http://spark.apache.org).

To enable faster data import and demonstrating `SQL`-related capabilities, the `.CSV` data files provided on Kaggle have been uploaded to a `PostgreSQL` database hosted on Amazon Web Services (AWS) Relational Database Service (RDS) under Chicago Booth Analytics's account.

We will also showcase visualization packages that are particularly well-integrated with data frames, especially __`GGplot2`__, although visualization is not the focus of this tutorial and will be covered more extensively in other tutorials.

Let's convert certain string columns to categorical factors, and convert `timestamp_first_active` to `POSIXlt` date-times:

```r
# let's create a custom parser of character strings to POSIXlt date-times
converter_from_str_to_POSIXlt <-
    function(from) {
        s <- regmatches(from, regexpr('[0-9]+', from, perl=TRUE))
        as.POSIXlt(
            paste(
                substr(s, 1, 4), '-', substr(s, 5, 6), '-', substr(s, 7, 8), ' ',
                substr(s, 9, 10), ':', substr(s, 11, 12), ':', substr(s, 13, 14), sep=''),
            format='%Y-%m-%d %H:%M:%S')
    }

setClass('POSIXlt_DateTime')
setAs('character', 'POSIXlt_DateTime', converter_from_str_to_POSIXlt)
```

```r
users_df$gender <- factor(users_df$gender)
users_df$signup_method <- factor(users_df$signup_method)
users_df$signup_flow <- factor(users_df$signup_flow)
users_df$language <- factor(users_df$language)
users_df$affiliate_channel <- factor(users_df$affiliate_channel)
users_df$affiliate_provider <- factor(users_df$affiliate_provider)
users_df$first_affiliate_tracked <- factor(users_df$first_affiliate_tracked)
users_df$signup_app <- factor(users_df$signup_app)
users_df$first_device_type <- factor(users_df$first_device_type)
users_df$first_browser <- factor(users_df$first_browser)
users_df$country_destination <- factor(users_df$country_destination)

users_df$timestamp_first_active <- as(users_df$timestamp_first_active, 'POSIXlt_DateTime')

cat("'users' data.frame column data types after conversion:")
sapply(users_df, class)

cat("'users' data.frame summary after conversion:")
summary(users_df)
```

There is currently no method to read directly from a `SQL` database to a `data.table`. If the original source is a `SQL` database, the only way to get a `data.table` is to convert from an intermediate `data.frame`:

```r
cat("Converting 'users' data.frame to data.table... ", end='')
tic <- proc.time()
users_dt <- as.data.table(users_df)
toc <- proc.time()
cat('done after', (toc - tic)['elapsed'], 'seconds.\n\n')
    
cat("'users' data.table column data types:")
sapply(users_dt, class)

cat("'users' data.table summary:")
summary(users_dt)

cat("'users' data.table: (", NB_EXAMPLES_TO_SHOW, ' first rows only)\n', sep='')
users_dt[1 : NB_EXAMPLES_TO_SHOW]
```

Below, we perform data-cleaning for the `users` table, converting `-unknown-` categorical levels that are supposed to be `NULL`, fixing the `age` variable and extracting the date-time sub-components such as year, month, day, day of week & hour of day out of date-time variables:

```r
sqldf("DROP TABLE IF EXISTS users_clean_temp")

sqldf(
    "SELECT
        id,
        date_account_created AS acct_crt_date,
        EXTRACT(YEAR FROM date_account_created) AS acct_crt_yr,
        EXTRACT(MONTH FROM date_account_created) AS acct_crt_mth,
        EXTRACT(DAY FROM date_account_created) AS acct_crt_day,
        EXTRACT(DoW FROM date_account_created) AS acct_crt_day_of_wk_idx,
        CASE
            WHEN date_account_created IS NULL THEN NULL
            ELSE CONCAT(EXTRACT(DoW FROM date_account_created), ': ',
                        TO_CHAR(date_account_created, 'Dy'))
            END
            AS acct_crt_day_of_wk,
        CAST(
            CONCAT(
                SUBSTR(timestamp_first_active, 1, 4),
                '-',
                SUBSTR(timestamp_first_active, 5, 2),
                '-',
                SUBSTR(timestamp_first_active, 7, 2),
                ' ',
                SUBSTR(timestamp_first_active, 9, 2),
                ':',
                SUBSTR(timestamp_first_active, 11, 2),
                ':',
                SUBSTR(timestamp_first_active, 13, 2))
            AS TIMESTAMP)
            AS fst_actv_ts,
        date_first_booking AS fst_bkg_date,
        EXTRACT(YEAR FROM date_first_booking) AS fst_bkg_yr,
        EXTRACT(MONTH FROM date_first_booking) AS fst_bkg_mth,
        EXTRACT(DAY FROM date_first_booking) AS fst_bkg_day,
        EXTRACT(DoW FROM date_first_booking) AS fst_bkg_day_of_wk_idx,
        CASE
            WHEN date_first_booking IS NULL THEN NULL
            ELSE CONCAT(EXTRACT(DoW FROM date_first_booking), ': ',
                        TO_CHAR(date_first_booking, 'Dy'))
            END
            AS fst_bkg_day_of_wk,
        CASE
            WHEN age > 1000 THEN 2015 - age
            ELSE age
            END
            AS age,
        CASE
            WHEN gender = '-unknown-' THEN NULL
            ELSE gender
            END
            AS gender,
        language,
        signup_method,
        signup_flow,
        signup_app,
        first_device_type,
        CASE
            WHEN first_browser = '-unknown-' THEN NULL
            ELSE first_browser
            END
            AS first_browser,
        affiliate_channel,
        affiliate_provider,
        CASE
            WHEN first_affiliate_tracked = 'null' THEN NULL
            ELSE first_affiliate_tracked
            END
            AS first_affiliate_tracked,
        country_destination
    INTO
        TEMP TABLE users_clean_temp
    FROM
        users")
```

To double-check that the categorical variables `gender`, `first_affiliate_tracked` and `first_browser` have been fixed:

```r
cat('Categorical Levels of Variables:\n\n')

for (v in c('gender', 'first_affiliate_tracked', 'first_browser')) {
    print(
        sqldf(
            paste(
                "SELECT
                    DISTINCT", v,
                "FROM
                    users_clean_temp
                WHERE",
                    v, "IS NOT NULL")))
    cat('\n')
}
```

## Generate `age_from` and `age_to` to define each `age_bucket`

### ___`PostgreSQL`___

Highlights:

- __`CASE ... WHEN ... THEN ... WHEN ... THEN ... ELSE ... END`__ statement in SQL; 
<br><br>

- __`CAST`__ function to convert among data types in SQL;
<br><br>

- __`REGEXP_MATCHES`__ function in PostgreSQL to extract a arrays of pattern-matching sub-strings from a string accordingly a regular expression.

```r
sqldf("DROP TABLE IF EXISTS age_gender_destination_stats_with_agefrom_and_ageto")

sqldf(
    paste(
        "SELECT
            age_bucket,
            CAST(age_from_to[1] AS DOUBLE PRECISION) - 0.5 AS age_from,
            CAST(age_from_to[2] AS DOUBLE PRECISION) + 0.5 AS age_to,
            gender,
            country_destination,
            population_in_thousands
        INTO
            TEMP TABLE age_gender_destination_stats_with_agefrom_and_ageto
        FROM

            (SELECT
                age_bucket,
                CASE
                    WHEN age_bucket = '100+' THEN ARRAY['100', '", max_age_cutoff, "']
                    ELSE REGEXP_MATCHES(age_bucket, '([0-9]*)-([0-9]*)')
                         -- alternatives in this case:
                         -- REGEXP_SPLIT(age_bucket, '-') or
                         -- STRING_TO_ARRAY(age_bucket, '-')
                    END
                    AS age_from_to,   -- note that this is an ARRAY
                gender,
                country_destination,
                population_in_thousands
            FROM
                age_gender_destination_stats) sub_query

        ORDER BY
            age_from,
            gender,
            country_destination",
        sep=''))

sqldf(
    "SELECT
        *
    FROM
        age_gender_destination_stats_with_agefrom_and_ageto")[1 : NB_EXAMPLES_TO_SHOW, ]
```

## Top 3 foreign first destinations by age bucket

### ___`PostgreSQL`___

Highlights:

- __`LEFT JOIN`__ in `SQL`;
<br><br>

- __Join Among More Than 2 Tables__ in `SQL`;
<br><br>

- __Non-Equality Joins__ in `SQL`;
<br><br>

- (Multiple, nested) __Sub-Queries__ in `SQL`;
<br><br>

- __Windowing__ (__`PARTITION BY`__, __`ORDER BY`__) in `SQL` / `HiveQL`, and __Windowed Analytics Functions__:
    - __`RANK() OVER`__

```r
sqldf(
    "SELECT
        sub_query_2.*,
        countries.destination_language AS dest_lang,
        countries.distance_km AS dist_km,
        countries.destination_km2 AS dest_area_km2
    FROM

        (SELECT
            age_bucket,
            RANK() OVER desc_nb_1st_bkgs_by_age_bucket AS dest_rank,
            dest,
            nb_1st_bkgs_2014,
            nb_1st_bkgs_2014 / SUM(nb_1st_bkgs_2014) OVER by_age_bucket AS prop_in_bucket_nb_bkgs,
            nb_1st_bkgs_2014 / nb_all_bkgs_2015 AS prop_in_all_bkgs
        FROM

            (SELECT
                age_buckets.age_bucket AS age_bucket,
                users_clean.country_destination AS dest,
                COUNT(DISTINCT users_clean.id) AS nb_1st_bkgs_2014,
                age_dest_stats.nb_all_bkgs_2015 AS nb_all_bkgs_2015
            FROM
                users_clean
                    LEFT JOIN age_buckets
                        ON users_clean.age BETWEEN age_buckets.age_from AND age_buckets.age_to
                    LEFT JOIN

                        (SELECT
                            age_bucket,
                            country_destination AS dest,
                            1000 * SUM(population_in_thousands) AS nb_all_bkgs_2015
                        FROM
                            age_gender_destination_stats
                        GROUP BY
                            age_bucket,
                            country_destination) age_dest_stats

                        ON users_clean.country_destination = age_dest_stats.dest AND
                           age_buckets.age_bucket = age_dest_stats.age_bucket

            WHERE
                users_clean.country_destination NOT IN ('US', 'other', 'NDF') AND
                users_clean.fst_bkg_yr = 2014
            GROUP BY
                age_buckets.age_bucket,
                users_clean.country_destination,
                nb_all_bkgs_2015) sub_query_1

        WINDOW
            by_age_bucket AS
                (PARTITION BY age_bucket),
            desc_nb_1st_bkgs_by_age_bucket AS
                (PARTITION BY age_bucket
                 ORDER BY nb_1st_bkgs_2014 DESC)) sub_query_2

        LEFT JOIN countries
            ON sub_query_2.dest = countries.country_destination
    WHERE
        sub_query_2.dest_rank <= 3
    ORDER BY
        age_bucket,
        dest_rank")[1 : NB_EXAMPLES_TO_SHOW, ]
```

Alternative soluton _without_ using windowing techniques:

```r
sqldf("DROP TABLE IF EXISTS temp_table_1")

sqldf(
    "SELECT
        age_buckets.age_bucket AS age_bucket,
        users_clean.country_destination AS dest,
        COUNT(DISTINCT users_clean.id) AS nb_1st_bkgs_2014,
        age_dest_stats.nb_all_bkgs_2015 AS nb_all_bkgs_2015
    INTO
        TEMP TABLE temp_table_1
    FROM
        users_clean
            LEFT JOIN age_buckets
                ON users_clean.age BETWEEN age_buckets.age_from AND age_buckets.age_to
            LEFT JOIN

                (SELECT
                    age_bucket,
                    country_destination AS dest,
                    1000 * SUM(population_in_thousands) AS nb_all_bkgs_2015
                FROM
                    age_gender_destination_stats
                GROUP BY
                  age_bucket,
                  country_destination) age_dest_stats

                ON users_clean.country_destination = age_dest_stats.dest AND
                   age_buckets.age_bucket = age_dest_stats.age_bucket
    WHERE
        users_clean.country_destination NOT IN ('US', 'other', 'NDF') AND
        users_clean.fst_bkg_yr = 2014
    GROUP BY
        age_buckets.age_bucket,
        users_clean.country_destination,
        nb_all_bkgs_2015
    ORDER BY
        age_bucket,
        nb_1st_bkgs_2014 DESC,
        nb_all_bkgs_2015 DESC")

cat('temp_table_1:\n')
sqldf(
    "SELECT
        *
    FROM
        temp_table_1")
```

```r
sqldf("DROP TABLE IF EXISTS nb_1st_bkgs_by_date")

sqldf(
    "SELECT
        fst_bkg_date AS date,
        COUNT(*) AS nb_1st_bkgs
    INTO
        TEMP TABLE nb_1st_bkgs_by_date
    FROM
        users_clean
    WHERE
        fst_bkg_date IS NOT NULL
    GROUP BY
        fst_bkg_date
    ORDER BY
        date")

cat('Number of first bookings by date: (first', NB_EXAMPLES_TO_SHOW, 'rows only)\n')
sqldf(
    "SELECT
        *
    FROM
        nb_1st_bkgs_by_date")[1 : NB_EXAMPLES_TO_SHOW, ]
```

```r
sqldf("DROP TABLE IF EXISTS nb_1st_bkgs_by_wk")

sqldf(
    "SELECT
        sunday_end_of_wk,
        SUM(nb_1st_bkgs) AS nb_1st_bkgs_in_wk
    INTO
        TEMP TABLE nb_1st_bkgs_by_wk
    FROM

        (SELECT
            CASE
                WHEN CAST(EXTRACT(DoW FROM date) AS INT) % 7 > 0
                    THEN date + 7 - CAST(EXTRACT(DoW FROM date) AS INT) % 7
                ELSE date
                END
                AS sunday_end_of_wk,
            nb_1st_bkgs
        FROM
            nb_1st_bkgs_by_date) sub_query
        
    GROUP BY
        sunday_end_of_wk
    ORDER BY
        sunday_end_of_wk")

sqldf(
    "SELECT
        *
    FROM
        nb_1st_bkgs_by_wk")[1 : NB_EXAMPLES_TO_SHOW, ]
```

### ___`PostgreSQL`___

Highlights:

- __`RIGHT JOIN`__: works similarly to `LEFT JOIN`;
<br><br>

- __`CROSS JOIN`__ to create Cartesian product;
<br><br>

- __`COALESCE(...)`__ for replacing `NULL` values in `SQL`;
<br><br>

- __`PARTITION BY`__ & __`ORDER BY`__ windowing in `SQL`;
<br><br>

- __`RANK() OVER`__ windowed function in `SQL`;
<br><br>

- __`LAG(...) OVER`__ windowed function in `SQL`;
<br><br>

- Multiple, nested __sub-queries__ in `SQL`.

```r
sqldf("DROP TABLE IF EXISTS nb_1st_bkgs_by_mth_by_dest")

sqldf(
    "SELECT
        CAST(DATE_TRUNC('month', date) AS DATE) AS start_of_mth,
        dest,
        SUM(nb_1st_bkgs) AS nb_1st_bkgs_in_mth
    INTO
        TEMP TABLE nb_1st_bkgs_by_mth_by_dest
    FROM

        (SELECT
            fst_bkg_date AS date,
            country_destination AS dest,
            COUNT(*) AS nb_1st_bkgs
        FROM
            users_clean
        WHERE
            fst_bkg_date IS NOT NULL AND
            country_destination NOT IN ('US', 'other', 'NDF')
        GROUP BY
            fst_bkg_date,
            country_destination) nb_1st_bkgs_by_date_by_dest

    GROUP BY
        DATE_TRUNC('month', date),
        dest
    ORDER BY
        start_of_mth,
        dest")

cat('Number of Foreign First Bookings by Month, by Destination: (first', NB_EXAMPLES_TO_SHOW, 'rows only)\n')
sqldf(
    "SELECT
        *
    FROM
        nb_1st_bkgs_by_mth_by_dest")[1 : NB_EXAMPLES_TO_SHOW, ]
```

```r
sqldf("DROP TABLE IF EXISTS nb_1st_bkgs_incl_zeros_by_mth_by_dest")

sqldf(
    "SELECT
        all_mths_and_foreign_dests.start_of_mth AS start_of_mth,
        all_mths_and_foreign_dests.dest AS dest,
        COALESCE(nb_1st_bkgs_by_mth_by_dest.nb_1st_bkgs_in_mth, 0) AS nb_1st_bkgs_in_mth
    INTO
        TEMP TABLE nb_1st_bkgs_incl_zeros_by_mth_by_dest
    FROM
        nb_1st_bkgs_by_mth_by_dest

        RIGHT JOIN

        (SELECT
            all_mths.start_of_mth AS start_of_mth,
            foreign_dests.column1 AS dest
        FROM

            (SELECT
                DISTINCT start_of_mth
            FROM
                nb_1st_bkgs_by_mth_by_dest) all_mths

            CROSS JOIN

            (VALUES ('AU'), ('CA'), ('DE'), ('ES'), ('FR'),
                    ('GB'), ('IT'), ('NL'), ('PT')) foreign_dests) all_mths_and_foreign_dests

        ON nb_1st_bkgs_by_mth_by_dest.start_of_mth = all_mths_and_foreign_dests.start_of_mth AND
           nb_1st_bkgs_by_mth_by_dest.dest = all_mths_and_foreign_dests.dest
    ORDER BY
        start_of_mth,
        dest")

cat('Augmenting the above table with zeros for Destinations with no first bookings in a Month:\n')
sqldf(
    "SELECT
        *
    FROM
        nb_1st_bkgs_incl_zeros_by_mth_by_dest")
```

```r
cat('Rankings among Destinations by Month:\n')

sqldf(
    "SELECT
        start_of_mth,
        dest_rank_in_mth,
        LAG(dest_rank_in_mth, 1, NULL) OVER inc_mth_by_dest - dest_rank_in_mth AS dest_rank_rise_vs_last_mth,
        dest,
        nb_1st_bkgs_in_mth
    FROM
        (SELECT
            start_of_mth,
            RANK() OVER desc_nb_1st_bkgs_by_mth AS dest_rank_in_mth,
            dest,
            nb_1st_bkgs_in_mth
        FROM
            nb_1st_bkgs_incl_zeros_by_mth_by_dest
        WINDOW
            desc_nb_1st_bkgs_by_mth AS
                (PARTITION BY start_of_mth
                 ORDER BY nb_1st_bkgs_in_mth DESC)) sub_query
    WINDOW
        inc_mth_by_dest AS
            (PARTITION BY dest
             ORDER BY start_of_mth)
    ORDER BY
        start_of_mth,
        dest_rank_in_mth")
```

Alternative solution for the last step _without_ using windowing techniques:

```r
sqldf("DROP TABLE IF EXISTS temp_table")

sqldf(
    "SELECT

        (SELECT
            COUNT(*) + 1
        FROM
            nb_1st_bkgs_incl_zeros_by_mth_by_dest clone
        WHERE
            clone.dest = nb_1st_bkgs_incl_zeros_by_mth_by_dest.dest AND
            clone.start_of_mth < nb_1st_bkgs_incl_zeros_by_mth_by_dest.start_of_mth)
        AS mth_order,

        nb_1st_bkgs_incl_zeros_by_mth_by_dest.start_of_mth AS start_of_mth,

        (SELECT
            COUNT(*) + 1
        FROM
            nb_1st_bkgs_incl_zeros_by_mth_by_dest clone
        WHERE
            clone.start_of_mth = nb_1st_bkgs_incl_zeros_by_mth_by_dest.start_of_mth AND
            clone.nb_1st_bkgs_in_mth > nb_1st_bkgs_incl_zeros_by_mth_by_dest.nb_1st_bkgs_in_mth)
        AS dest_rank_in_mth,

        nb_1st_bkgs_incl_zeros_by_mth_by_dest.dest AS dest,
        nb_1st_bkgs_incl_zeros_by_mth_by_dest.nb_1st_bkgs_in_mth AS nb_1st_bkgs_in_mth
    INTO
        TEMP TABLE temp_table
    FROM
        nb_1st_bkgs_incl_zeros_by_mth_by_dest")

cat('temp_table:\n')
sqldf(
    "SELECT
        *
    FROM
        temp_table")
```

```r
cat('results:\n')

sqldf(
    "SELECT
        temp_table.start_of_mth AS start_of_mth,
        temp_table.dest_rank_in_mth AS dest_rank_in_mth,
        temp_table_clone.dest_rank_in_mth - temp_table.dest_rank_in_mth
          AS dest_rank_rise_vs_last_mth,
        temp_table.dest AS dest,
        temp_table.nb_1st_bkgs_in_mth AS nb_1st_bkgs_in_mth
    FROM
        temp_table
            LEFT JOIN temp_table temp_table_clone
                ON temp_table.dest = temp_table_clone.dest AND
                   temp_table.mth_order = temp_table_clone.mth_order + 1
    ORDER BY
        start_of_mth,
        dest_rank_in_mth")
```
